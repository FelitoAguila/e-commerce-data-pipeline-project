4.1 Do you see any relationship between the Brazilian public holidays of 2017 and that year's purchases?

The chart shows a limited relationship between Brazilian public holidays and purchase volumes. 

Most national holidayssuch as Tiradentes Day (April 21), Labor Day (May 1), Independence Day (September 7), Our Lady of Aparecida (October 12), All Souls' Day (November 2), and Republic Proclamation Day (November 15)align with the vertical dashed lines and correspond to periods of relatively stable or even slightly lower order activity, typically in the 80200 range per day, with no significant spikes. Carnival (late February), Good Friday/Easter (mid-April), and Corpus Christi (June) also show no notable increases in online purchases. 

The only dramatic surge occurs in late November to early December, peaking above 1,1001,200 orders around the end of November, followed by a sharp decline into January 2018. This massive peak aligns closely with Black Friday on November 24, 2017, and the subsequent Christmas shopping season, which are major commercial drivers in Brazilian e-commerce rather than traditional public holidays. 

Overall, while proximity to Christmas contributes indirectly, standard public holidays do not appear to strongly boost online orders in this datasetlikely because many Brazilians use these days for leisure, travel, or in-person spending instead of e-commerce.
